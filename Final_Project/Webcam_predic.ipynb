```python
import numpy as np
from keras import models
import time
import cv2
import tensorflow as tf
from PIL import Image, ImageDraw, ImageFont
```

```python
import numpy as np
decoder = {'0' : 'Downward-Facing Dog', 
           '1' : 'Upward-Facing Dog',  
           '2' : 'Warrior One Pose', 
           '3' : 'Warrior Two Pose', 
           '4' : 'Warrior Three Pose'}
```

```python
name='MobileNet32_v1.h5'
model = models.load_model(name)
print("load ",name)

# 選擇第攝影機
cap = cv2.VideoCapture(0)
```

```python
model.summary()
```

```python
while(True):
    # 從攝影機擷取一張影像

    ret, frame = cap.read()
    # print(frame)
    
    # 模型預估
    frame_resize = cv2.resize(frame, (32, 32), interpolation=cv2.INTER_AREA)
    input_data=frame_resize[np.newaxis,:]
    input_data = input_data/255
    # print(input_data.shape)
    start = time.time()
    
    predict_prob = model.predict(input_data)
    predicted_labels = np.argmax(predict_prob, axis=1)
    print(predicted_labels)
    
    end = time.time()
    accumulate_time = (end - start)
    FPS = 1/accumulate_time
    FPS = round(FPS,2)

    result = decoder[str(predicted_labels[0])]
    
    # print(f"result = {result}")
    # detected = translate_Y(result[0], show=True)
    #print(detected)
    
    
    # 顯示圖片
    cv2.putText(frame, str(FPS), (540, 40), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 1, cv2.LINE_AA)
    cv2.putText(frame, result, (10, 40), cv2.FONT_HERSHEY_SIMPLEX,1, (255, 0, 0), 1, cv2.LINE_AA)
    cv2.imshow('frame', frame)
        
        # 若按下 q 鍵則離開迴圈
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 釋放攝影機
cap.release()

# 關閉所有 OpenCV 視窗
cv2.destroyAllWindows()
```
