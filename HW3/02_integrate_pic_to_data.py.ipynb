```python
import scipy.io as sio 
import numpy as np
import cv2
import os
```

```python
train_file_path = 'train'
all_file_name = os.listdir(train_file_path)
print(all_file_name)
Y_data = []
```

```python
# rotate image
for i in range(len(all_file_name)):
    pic_file_path = train_file_path + '/' + all_file_name[i] 
    all_pic_name = os.listdir(pic_file_path)
    for j in range(len(all_pic_name)):
        pic_path = pic_file_path + '/' + all_pic_name[j]
        img = cv2.imread(pic_path)
        output_ROTATE_90_CLOCKWISE = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
        output_ROTATE_90_COUNTERCLOCKWISE = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
        output_ROTATE_180 = cv2.rotate(img, cv2.ROTATE_180)
        cv2.imwrite(pic_path+f'rotate90_{j}.jpg', output_ROTATE_90_CLOCKWISE)
        cv2.imwrite(pic_path+f'rotate270_{j}.jpg', output_ROTATE_90_COUNTERCLOCKWISE)
        cv2.imwrite(pic_path+f'rotate180_{j}.jpg', output_ROTATE_180)
```

```python
count = 0
for i in range(len(all_file_name)):
    pic_file_path = train_file_path + '/' + all_file_name[i] 
    all_pic_name = os.listdir(pic_file_path)
    for j in range(len(all_pic_name)):
        pic_path = pic_file_path + '/' + all_pic_name[j]
        img = cv2.imread(pic_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        ret, img = cv2.threshold(img, 220, 255, cv2.THRESH_BINARY)
        img = cv2.resize(img, (32, 32), interpolation=cv2.INTER_AREA)
#         print("resize size:", img.shape)
        #
        if count == 0:
            X_data = img
            print(X_data.shape)
            X_data = np.expand_dims(X_data, axis = 0) 
            X_data = np.expand_dims(X_data, axis = 3)
            print("expand size:", X_data.shape)
        else:
            X_data = np.vstack((X_data,img[np.newaxis,:,:,np.newaxis]))
            print("stack size:", X_data.shape)
        Y_data.append(i) #label
        print(str(i) + '-' + str(j)) #
        count += 1                   
Y_data = np.array(Y_data)

save_train = 'bw_v2.mat'
save_array_x = X_data
save_array_y = Y_data
sio.savemat(save_train, {'X_data': save_array_x, 'Y_data': save_array_y})    
```
