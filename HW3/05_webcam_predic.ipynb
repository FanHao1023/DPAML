```python
import numpy as np
from keras import models
import time
import cv2
import tensorflow as tf
```

```python
import numpy as np

def one_hot_decode(data):
    # 查找 one-hot 編碼中值為最大的索引
    max_index = np.argmax(data)
    
    # 獲取 one-hot 編碼中值為最大的數值
    max_value = np.max(data)
    
    # 返回最大值的索引和數值
    return int(max_index), float(max_value)



def translate_Y(yi , show = False):
    
    num = {
        0 : "call", 1 : "mute", 2 : "paper", 3 : 'rock', 4 : 'rock_and_roll', 5 : 'scissors'} 
    
    yi_new = num.get( yi ) 
            
    if show:
        print(yi)
        print(yi_new)            
        
    return yi_new
```

```python
name='ResNet50_messy6_v2.h5'
model = models.load_model(name)
print("load ",name)

# 選擇第攝影機
cap = cv2.VideoCapture(0)
```

```python
model.summary()
```

```python
while(True):
    # 從攝影機擷取一張影像

    ret, frame = cap.read()
    frame = frame/255
    # print(frame)
    
    # 模型預估
    frame_resize = cv2.resize(frame, (224, 224), interpolation=cv2.INTER_AREA)
    input_data=frame_resize[np.newaxis,:]
    # print(input_data.shape)
    start = time.time()
    prediction=model.predict(input_data) 
    end = time.time()
    accumulate_time = (end - start)
    FPS = 1/accumulate_time
    FPS = round(FPS,2)
    result = one_hot_decode(prediction)
    # print(f"result = {result}")
    detected = translate_Y(result[0], show=True)
    #print(detected)
    
    
    # 顯示圖片
    cv2.putText(frame, str(FPS), (540, 40), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 1, cv2.LINE_AA)
    cv2.putText(frame, detected, (10, 40), cv2.FONT_HERSHEY_SIMPLEX,1, (255, 0, 0), 1, cv2.LINE_AA)
    cv2.imshow('frame', frame)
        
        # 若按下 q 鍵則離開迴圈
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 釋放攝影機
cap.release()

# 關閉所有 OpenCV 視窗
cv2.destroyAllWindows()
```
